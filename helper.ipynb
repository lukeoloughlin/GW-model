```python
def create_getter_setter(name):
    return f"""
    @property
    def {name}(self) -> float:
        return self.__cpp_struct.{name}

    @{name}.setter
    def {name}(self, value: float) -> None:
        assert_positive(value, '{name}')
        self.__cpp_struct.{name} = value
    """

def create_parameter_interface(names):
    out_str = create_getter_setter(names[0])
    for name in names[1:]:
        out_str += create_getter_setter(name)
    return out_str
```

```python
print(create_parameter_interface(['T', 'CSA']))
```
